```python
# Importing pandas
import pandas as pd
#import matplotlib
import matplotlib.pyplot as PLT

import numpy as np
#open the file
df = pd.read_csv("NYCOpenData.csv")

#testing dataset
df.head()
```

1) What is the average Base Salary for Police Officers from 2015 to 2023?

```python
#only the Police Department is the Agency we need
PoliceDF = df[df['Agency Name'] == 'POLICE DEPARTMENT']

#want the salary of police officers per year
PoliceDF = PoliceDF[df['Pay Basis'] == 'per Annum']

#want the average base salary
Pay = PoliceDF['Base Salary'].mean()

#average salary of police officers per year
AsOf = PoliceDF.groupby(['Fiscal Year'])['Base Salary'].mean()

PLT.ylabel("Base Salary /Annum")
#plot the graph

#change background color to dark
PLT.style.use('dark_background')
line = AsOf.plot(label='Sine Wave', color='lightblue', linewidth=3, linestyle='-', marker='o', markersize=8)
PLT.title("Average Base Salary From 2015-2023")
AsOf.plot()
```
