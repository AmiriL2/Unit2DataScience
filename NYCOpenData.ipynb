```python
# Importing pandas
import pandas as pd
#import matplotlib
import matplotlib.pyplot as PLT

import numpy as np
#open the file
df = pd.read_csv("NYCOpenData.csv")

#testing dataset
df.head()
```

1) What is the average Base Salary for Police Officers from 2015 to 2023?

```python
#only the Police Department is the Agency we need
PoliceDF = df[df['Agency Name'] == 'POLICE DEPARTMENT']

#want the salary of police officers per year
PoliceDF = PoliceDF[df['Pay Basis'] == 'per Annum']

#want the average base salary
Pay = PoliceDF['Base Salary'].mean()

#average salary of police officers per year
AsOf = PoliceDF.groupby(['Fiscal Year'])['Base Salary'].mean()

PLT.ylabel("Base Salary /Annum")
#plot the graph

#change background color to dark
PLT.style.use('dark_background')
line = AsOf.plot(label='Sine Wave', color='lightblue', linewidth=3, linestyle='-', marker='o', markersize=8)
PLT.title("Average Base Salary From 2015-2023")
AsOf.plot()
```

Based on the information displayed on this graph, it can be said that throughout the years there has been a massive increase in the average base salary of police officers, growing from slightly above $68,000/year to around $82,000/year. This is a growth of at minimum 20% from 2015 to 2023.

2) What has been the Average Salary Police Officers throughout the different boroughs?

```python
#only the Police Department is the Agency we need
PoliceDF = df[(df['Agency Name'] == 'POLICE DEPARTMENT') & (df['Pay Basis'] == 'per Annum')]

#groups the average by the average salary of officers throughout the 5 boroughs
Average = PoliceDF.groupby(['Work Location Borough'])['Base Salary'].mean()

#changes background colors 
with PLT.style.context("rose-pine-moon.mplstyle"):
    Average.plot(kind='bar', rot= 45)
PLT.title("Average Base Salary Based on Borough")
```

Based on what is displayed from this graph, it can be said that Manhattan has the highest base salary per annum for police officers, with an average of around $80,000 per year while the lowest borough is Brooklyn with just under $70,000 per year.

3) What is the average Base Salary for Teachers from 2014 to 2023?

```python
#only the Admin for child services is the Agency we need
TeachersDF = df[df['Agency Name'] == "ADMIN FOR CHILDREN'S SVCS"]

#want the salary of teachers per year
TeachersDF = TeachersDF[df['Pay Basis'] == 'per Annum']

#want the average base salary
Pay = TeachersDF['Base Salary'].mean()

#average salary of teachers per year
AsOf = TeachersDF.groupby(['Fiscal Year'])['Base Salary'].mean()

#label for y-axis
PLT.ylabel("Base Salary /Annum")

#change background color to dark
PLT.style.use('dark_background')

#used to plot the graph, label it, change the color, width, and style
line = AsOf.plot(label='Sine Wave', color='lightblue', linewidth=3, linestyle='-', marker='o', markersize=8)
PLT.title("Average Base Salary From 2014-2023")
```

Based on the data shown in the chart above, it can be said that there has been a steady increase of teachers base salary by at least  17% as it has grown from just slightly under $60,000/year to a little over $70,000/year
