*Use the code block below to open `sales.csv` with python. Then create a variable called `content` which is equal to a list of each line of the file*

````python
SaleFile = open('sales.csv' , 'r')
with open("sales.csv" , "r") as SaleFile:
    content = SaleFile.readlines()
    print(content)
````

*Using the `split` function and a `for` loop calculate the total sales in a month*

*Hint: you will need a seperate variable for each month*

````python
#all the 3 months that are on the list
Jan = 0
Feb = 0
Mar = 0

#delete the first line cause that is a string
del content[0]
sales = open("sales.csv")

#for loop to acess specific content
for sales in content:
    this = sales.split(",")
    #third index is the total
    thisDate = this[3]
    #the "/" represents the month
    month = thisDate.split("/")[0]
    #float so that all the total sales is printed
    total = float(this[2])

    #if the month is Jan
    if month == "1":
        #add that months total to Jan
        Jan += total
         #if the month is Fed
    elif month == "2":
        #add that months total to Feb
        Feb += total
         #if the month is Mar
    elif month == "3":
        #add that months total to Mar
        Mar += total
   
   #print the sales for those months
print(Jan)
print(Feb)
print(Mar)
   
````

*Using your calculated total sales create a graph to represent this data and write a written analysis*

````python
import matplotlib.pyplot as PLT 
y = [Jan, Feb, Mar]
x = ["January", "Febuary", "March"]

#plot them on a graoh
PLT.plot(x,y)
PLT.xlabel("Months")
PLT.ylabel("Total Sales")
#this is so that it goes from 0 to 55,000 on the y axis
PLT.ylim(0,55000)
````

The data presented in the graph reveals that through Q1 of 2019, March had the best income with $52615.75. January comes second with $48622.69 and March is last with an income of $42499.60

*Using the `split` function, seperate each row of data. Then append the customer's name to a list called `customers`*

````python
#create an empty list for the customers
Customer = []
#for values in the content
for x in content:

    x = x.strip().split(",")
    #customer is the first index
    CustomerName = x[0]
    #add all the customer names to the list
    Customer.append(CustomerName)
    #print out their names
print(Customer)
````
